```
import numpy as np
```

```
inputs = np.array([[0,0,1],[1,0,1],[1,1,0],[0,1,0]])
inputs
```

```
outputs = np.array([[0],[1],[1],[0]])
outputs
```

```
np.random.seed(10)
weights = np.random.random((3,1))  ## random number b/w 0-1
weights
```

```
sum = np.dot(inputs,weights) + 0.02  ### 0.02-bias
sum
```

```
## Activation Function
def activation(x):
  return 1/(1+np.exp(-x))
```

```
pred_op = activation(sum)
pred_op
```

```
pred_op.round()
```

```
### Calculate Error
sum = np.dot(inputs,weights) + 0.02
pred_op = activation(sum)
error = outputs - pred_op
error
```

```
def gradient(x,a):
  return a * x*(1-x)
```

```
sum = np.dot(inputs,weights) + 0.02
pred_op = activation(sum)
error = outputs - pred_op
cng_in_wt = gradient(pred_op,error)  ## gradient descent
weights = weights + np.dot(inputs.T,cng_in_wt)  ## back propogation
error
```

```
### Training of network
for i in range(500):
  sum = np.dot(inputs,weights) + 0.02
  pred_op = activation(sum)
  error = outputs - pred_op
  cng_in_wt = gradient(pred_op,error)  ## gradient descent
  weights = weights + np.dot(inputs.T,cng_in_wt)  ## back propogation
```

```
import pandas as pd
```

```
data = pd.read_csv("/content/breast_cancer.csv")
data.head()
```

```
data.info()
```

```
data.isnull().sum()
```

```
data.columns
```

```
data.drop(["Unnamed: 32"],axis=1,inplace=True)
```

```
data.drop(["id"],axis=1,inplace=True)
```

```
data.isnull().sum()
```

```
from sklearn.preprocessing import LabelEncoder  ## the library for encoding
le = LabelEncoder()   ## object
data['diagnosis'] = le.fit_transform(data['diagnosis'])
# data['diagnosis'] = data['diagnosis'].map({""})
```

```
data.head(1)
```

```
### Divide the data into features and target
x = data.iloc[:,1:]
y = data.iloc[:,0]
y
```

```
y.value_counts()
```

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.80,
                                             random_state=1)
xtest
```

```
ytrain.value_counts()
```

```
len(ytrain)
x.info
```

```
## Apply the neural network from library
import tensorflow.keras as tf

## Create the blank model of Neural Network
model = tf.models.Sequential()

## add the layers
model.add(tf.layers.Dense(16,input_dim=30,activation='relu'))  ##input --> hidden
model.add(tf.layers.Dense(32,activation='relu'))  ## next hidden layer
model.add(tf.layers.Dense(1,activation='sigmoid')) ##output layer

## compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ["accuracy"])
```

```
### Train the model
model.fit(xtrain,ytrain,epochs=500)
```

```
### Predictions from the testing data
ypred = model.predict(xtest)
ypred = ypred.round()
ypred
```

```
## print actual value
ytest
```

```
model.evaluate(xtest,ytest)
```

```
### Confusion matrix - Square matrix of size (no of class) (2x2)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```
## to find the wrong predictions and their positions
id = []
for i in range(len(ytest)):
  if ytest.iloc[i]!=ypred[i]:
    id.append(i)
id
```
